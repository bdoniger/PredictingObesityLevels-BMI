```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = ['Age', 'Height_meters', 'Weight_kg']

plt.figure(figsize=(12, 4))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, len(numeric_cols), i)
    sns.boxplot(y=obesity[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()

for col in numeric_cols:
    Q1 = obesity[col].quantile(0.25)
    Q3 = obesity[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    outliers = obesity[(obesity[col] < lower) | (obesity[col] > upper)]
    print(f"{col}: {len(outliers)} potential outliers")
```

```python
target = 'Obesity_Level'
class_counts = obesity[target].value_counts()
class_percent = obesity[target].value_counts(normalize=True) * 100

print("Class counts:\n", class_counts)
print("\nClass percentages:\n", class_percent)

# Optional visualization
plt.figure(figsize=(8,4))
sns.countplot(x=target, data=obesity, order=class_counts.index)
plt.xticks(rotation=45)
plt.title('Obesity Level Distribution')
plt.show()
```
